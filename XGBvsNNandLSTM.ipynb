```python
matrix = []
for i in range(len(data['Close'])-1, 4, -1):
  day5=data['Close'][i-5]
  day4=data['Close'][i-4]
  day3=data['Close'][i-3]
  day2=data['Close'][i-2]
  day1=data['Close'][i-1]
  matrix.append([data['Close'][i], day1, day2, day3, day4, day5])

matrix
```

```python
df = pd.DataFrame(matrix,columns=['Close','One day ago','Two days ago','Three days ago','Four days ago', 'Five days ago'])
```

```python
df
```
