## Bubble charts

На следниот график можат да се видат вредностите во зависност од фреквенција на користење на виртуелниот простор, колку се олеснети секојдневните активности, колку пријатели имаат и колку социјална поддршка добиваат од тие пријатели

```python
fig = px.scatter(data, x="Многу секојдневни активности ми се олеснети благодарение на социјалните мрежи",
                 y="Имам пријатели на социјалните мрежи.",
                 color="Колку често на ден користите социјалните мрежи моментално?",
                 size='Добивам социјална поддршка од пријателите на социјалните мрежи.',
                 size_max=60)
fig.show()
```

## Pie charts

Графикот ни покажува какви се резултатите за социјалните интеракции доколку исклучиме еден од одговорите. На пример ако ги исклучиме воздржаните(Неутрално) ќе заклучиме дека 54.5% од испитаниците не се сложуваат дека виртуелниот простор ќе им ја намали осаменоста

```python
fig = px.pie(data, names='Користењето на виртуелниот простор ќе  [ми ја намали осаменоста]')
fig.show()
```

```python
fig = px.pie(data, names='Користењето на виртуелниот простор ќе  [ми ги подобри социјалните интеракции.]')
fig.show()
```

Како што интерпретиравме во нашето истражување виртуелниот простор покажува мешани резултати кога станува збор за подобрување на квалитетот на животот, односно извршувањето на професионалните обврски се олеснува додека пак за социјалните интеракции се чувствува одбивност према социјалните онлајн платформи.

```python
fig = px.pie(data, names='Користењето на виртуелниот простор ќе  [ми ги олеснува работните/студиските обврски]')
fig.show()
```

## Bar charts
