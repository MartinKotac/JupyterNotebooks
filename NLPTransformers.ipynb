```python
data = pd.read_csv('dataset.csv')
```

```python
data
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased',num_labels =2)
```

```python
train = data[data['data_type']=='train'].reset_index().drop(['index','data_type'],axis=1)
```

```python
primer = data[data['data_type']=='train'].reset_index()
primer.drop(['index','data_type'],axis=1,inplace=True)
primer
```

```python
test = data[data['data_type']=='val'].reset_index().drop(['index','data_type'],axis=1)
```

```python
train.to_csv('train.csv',index=False)
```

```python
test.to_csv('test.csv',index=False)
```

```python
dataset = load_dataset("csv", data_files={"train": "train.csv", "test": "test.csv"})
```

```python
dataset
```

```python
encoded_dataset = dataset.map(lambda t: tokenizer(t['abstract'],  truncation=True), batched=True,load_from_cache_file=False)
```

```python
from transformers import TrainingArguments, Trainer
```

```python
args = TrainingArguments(
    "sentiment",
    learning_rate=5e-5,
    num_train_epochs=3,
    per_device_eval_batch_size=12,
    per_device_train_batch_size=12,
    seed=19
)
```

```python
trainer = Trainer(
    model=model,
    args=args,
    tokenizer=tokenizer,
    train_dataset=encoded_dataset['train'],
    eval_dataset=encoded_dataset['test']
)
```

```python
trainer.train()
```

```python
y_pred = trainer.predict(encoded_dataset['test'])
```

```python
y_pred = y_pred.predictions
```

```python
import numpy as np
```

```python
y_pred = [np.argmax(y_pred[i]) for i in range(0,len(y_pred))]
```

```python
from sklearn.metrics import classification_report,f1_score
```

```python
print(classification_report(test['label'],y_pred))
```
